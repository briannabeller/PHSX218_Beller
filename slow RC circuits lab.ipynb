```python
#imports

import numpy as np
import matplotlib.pyplot as plt
```

```python
#data

v_charging = np.array([0,3.82,6.02,7.75,8.87,9.61,10.17,10.52,
                        10.76,10.93,11.03,11.11,11.16])

v_discharging = np.array([12.02,8.63,5.84,4.02,2.74,1.9,1.34,0.92,
                          0.66,0.47,0.33,0.23,0.17])

t = np.arange(0,125,10)
```

```python
#charging plot

plt.xlabel("Time (in s)")
plt.ylabel("Voltage (in V)")
plt.title("CHARGING PLOT")

plt.plot(t,v_charging)
```

```python
#discharging plot

plt.xlabel("Time (in s)")
plt.ylabel("Voltage (in V)")
plt.title("DISCHARGING PLOT")

plt.plot(t,v_discharging)
```
